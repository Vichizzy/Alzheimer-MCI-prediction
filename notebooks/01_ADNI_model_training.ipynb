```python
# ============================================================
# STEP 1, BLOCK 1 — Setup and load the diagnosis file
# ============================================================
# WHY: DXSUM is the foundation of everything. Before we touch
# any feature, we need to know which patients are MCI at
# baseline and what happens to them over time. This file
# defines our entire cohort.

import pandas as pd
import numpy as np
import os

# This is the folder containing your CONVERTED CSV files
# (the ones produced by the .rda to CSV conversion earlier)
csv_folder = r"C:\Users\Chima\Downloads\ADNIMERGE2\ADNIMERGE2\data_csv"

# Load the diagnosis summary file
# low_memory=False tells pandas to read the whole file before
# deciding column types — prevents incorrect type guessing on
# large files with mixed content
dxsum = pd.read_csv(os.path.join(csv_folder, 'DXSUM.csv'),
                    low_memory=False)

print(f"DXSUM loaded: {len(dxsum):,} rows")
print(f"Unique patients: {dxsum['RID'].nunique():,}")
print(f"\nColumns available:")
print(dxsum.columns.tolist())
```

```python
# ============================================================
# STEP 1, BLOCK 2 — Understand the DIAGNOSIS column
# ============================================================
# WHY: Before building any label, we must know exactly how
# diagnosis is coded in YOUR file. We expect 1=CN (normal),
# 2=MCI, 3=Dementia — but we confirm rather than assume.
# value_counts() shows how many rows have each value.
# dropna=False ensures we also see how many are missing.

print("DIAGNOSIS value distribution (all visits):")
print(dxsum['DIAGNOSIS'].value_counts(dropna=False).sort_index())
print("\nExpected coding: 1 = Cognitively Normal, 2 = MCI, 3 = Dementia")
```

```python
# ============================================================
# STEP 1, BLOCK 4 — Find patients who are MCI at baseline
# ============================================================
# WHY: Our entire prediction question is "for a patient who is
# MCI today, will they convert to Alzheimer's within 24 months?"
# So the first filter is: keep only patients whose BASELINE
# diagnosis is MCI. Everyone else is outside our question.

# Step A: isolate just the baseline visits
# We filter the full dxsum table to rows where VISCODE2 == 'bl'
baseline_visits = dxsum[dxsum['VISCODE2'] == 'bl'].copy()
print(f"Total baseline visits: {len(baseline_visits):,}")

# Step B: among those baseline rows, keep only MCI patients
# DIAGNOSIS is text in your data, so we match the string 'MCI'
mci_baseline = baseline_visits[baseline_visits['DIAGNOSIS'] == 'MCI'].copy()
print(f"Patients who are MCI at baseline: {len(mci_baseline):,}")

# Step C: confirm one row per patient (no duplicates at baseline)
# nunique() counts distinct RIDs; it should equal the row count
print(f"Unique MCI patients: {mci_baseline['RID'].nunique():,}")

# Step D: store the list of MCI patient IDs and their baseline dates
# We will need the baseline date to measure the 24-month window
mci_patients = mci_baseline[['RID', 'EXAMDATE']].copy()
mci_patients = mci_patients.rename(columns={'EXAMDATE': 'baseline_date'})
print(f"\nCohort defined: {len(mci_patients):,} MCI patients with baseline dates")
print(mci_patients.head())
```

```python
# ============================================================
# STEP 1, BLOCK 3 — Convert dates and sort visits in order
# ============================================================
# WHY: To track each patient over time, we need their visits
# in chronological order. EXAMDATE is currently text — we
# convert it to a real date so Python can sort and compare them.
# pd.to_datetime() turns "2011-05-14" style text into a date
# object that supports time calculations.

# Convert the exam date column from text to actual dates
# errors='coerce' means any unreadable date becomes NaT (missing)
# rather than crashing the whole operation
dxsum['EXAMDATE'] = pd.to_datetime(dxsum['EXAMDATE'], errors='coerce')

# Check how many dates failed to convert (should be very few)
print(f"Rows with unreadable dates: {dxsum['EXAMDATE'].isnull().sum()}")

# Sort the entire dataframe by patient, then by date within
# each patient. This puts every patient's visits in time order —
# essential for the label logic that scans forward in time.
dxsum = dxsum.sort_values(['RID', 'EXAMDATE']).reset_index(drop=True)

print("\nDates converted and visits sorted chronologically.")
print("\nExample — first patient's visit sequence:")
# Show one patient's visits to confirm the structure looks right
first_rid = dxsum['RID'].iloc[0]
print(dxsum[dxsum['RID'] == first_rid][['RID', 'VISCODE2', 'EXAMDATE', 'DIAGNOSIS']])
```

```python
# ============================================================
# STEP 1, BLOCK 5 — Build the conversion label
# ============================================================
# WHY: For each of our 1,328 MCI patients, we now look forward
# in time to answer one question: did they convert to Dementia
# within 24 months of their baseline visit?
#
# THE LOGIC:
#   - converter (1): DIAGNOSIS becomes "Dementia" at any visit
#                    within 24 months of baseline
#   - stable (0):    remained CN or MCI through a visit at/near
#                    the 24-month mark (had adequate follow-up)
#   - exclude:       no follow-up visit reaching ~24 months, so
#                    we cannot know their true outcome

# We define the conversion window as 24 months.
# We allow a small grace buffer (up to 30 months) so that a
# visit scheduled slightly late still counts — real visits
# rarely land exactly on the 24-month day.
WINDOW_MONTHS = 24
GRACE_MONTHS  = 30   # visits up to 30 months still count toward the window

# We'll build a list of results, one entry per patient
results = []

# Loop through each MCI patient one at a time
# iterrows() lets us step through the cohort row by row
for _, patient in mci_patients.iterrows():
    rid           = patient['RID']
    baseline_date = patient['baseline_date']

    # Skip patients whose baseline date failed to parse earlier —
    # without a valid baseline date we cannot measure the window
    if pd.isnull(baseline_date):
        continue

    # Get ALL of this patient's visits from the full dxsum table,
    # already sorted chronologically from Block 3
    patient_visits = dxsum[dxsum['RID'] == rid].copy()

    # Calculate months elapsed between baseline and each visit
    # (visit_date - baseline_date) gives a time difference;
    # dividing by ~30.44 days converts it to approximate months
    patient_visits['months_from_bl'] = (
        (patient_visits['EXAMDATE'] - baseline_date).dt.days / 30.44
    )

    # Keep only follow-up visits that fall AFTER baseline and
    # within our grace window (0 to 30 months). We exclude the
    # baseline visit itself (months_from_bl > 0).
    window_visits = patient_visits[
        (patient_visits['months_from_bl'] > 0) &
        (patient_visits['months_from_bl'] <= GRACE_MONTHS)
    ]

    # --- Determine the label ---

    # Did they convert? Check if "Dementia" appears in any
    # within-window visit's DIAGNOSIS
    converted = (window_visits['DIAGNOSIS'] == 'Dementia').any()

    # How far out does their follow-up actually reach?
    # We need at least one visit near the 24-month mark to
    # confidently call someone "stable"
    max_followup = window_visits['months_from_bl'].max() \
                   if len(window_visits) > 0 else 0

    if converted:
        # Find the month of FIRST conversion to Dementia
        # (useful metadata even though the label is binary)
        conv_visits = window_visits[window_visits['DIAGNOSIS'] == 'Dementia']
        conv_month  = conv_visits['months_from_bl'].min()
        results.append({
            'RID': rid,
            'baseline_date': baseline_date,
            'label': 1,
            'conversion_month': round(conv_month, 1),
            'max_followup_months': round(max_followup, 1)
        })
    elif max_followup >= (WINDOW_MONTHS - 6):
        # Stable: no conversion AND follow-up reaches at least
        # 18 months (24 minus a 6-month tolerance). We accept
        # 18+ months as adequate evidence of stability.
        results.append({
            'RID': rid,
            'baseline_date': baseline_date,
            'label': 0,
            'conversion_month': None,
            'max_followup_months': round(max_followup, 1)
        })
    else:
        # Insufficient follow-up — we cannot know their outcome,
        # so we exclude them by labeling as None
        results.append({
            'RID': rid,
            'baseline_date': baseline_date,
            'label': None,
            'conversion_month': None,
            'max_followup_months': round(max_followup, 1)
        })

# Turn the list of results into a dataframe
cohort = pd.DataFrame(results)

print(f"Total MCI patients processed: {len(cohort):,}")
print(f"\nLabel breakdown:")
print(cohort['label'].value_counts(dropna=False))
print(f"\n  1.0 = converter, 0.0 = stable, NaN = excluded (insufficient follow-up)")
```

```python
# ============================================================
# STEP 1, BLOCK 6 — Sanity check the label logic
# ============================================================
# WHY: Before trusting these labels, we manually inspect a few
# patients to confirm the code did what we intended. This is
# how you catch silent logic errors that no error message
# would reveal.

# Look at 3 converters — confirm they actually show Dementia
print("=== CHECKING 3 CONVERTERS ===")
converter_rids = cohort[cohort['label'] == 1]['RID'].head(3).tolist()
for rid in converter_rids:
    print(f"\nPatient RID {rid}:")
    visits = dxsum[dxsum['RID'] == rid][['VISCODE2', 'EXAMDATE', 'DIAGNOSIS']]
    print(visits.to_string(index=False))

# Look at 3 stable patients — confirm they stayed MCI/CN
print("\n\n=== CHECKING 3 STABLE PATIENTS ===")
stable_rids = cohort[cohort['label'] == 0]['RID'].head(3).tolist()
for rid in stable_rids:
    print(f"\nPatient RID {rid}:")
    visits = dxsum[dxsum['RID'] == rid][['VISCODE2', 'EXAMDATE', 'DIAGNOSIS']]
    print(visits.to_string(index=False))

# ============================================================
# STEP 1, BLOCK 7 — Examine conversion timing
# ============================================================
# WHY: Understanding WHEN patients convert within the 24-month
# window is useful clinical context and another validity check.
# If conversions cluster sensibly across the window rather than
# all at one suspicious point, the logic is behaving correctly.

print("Conversion timing (months from baseline):")
print(cohort[cohort['label'] == 1]['conversion_month'].describe())

# Also save the cohort so we never have to rebuild it
# index=False keeps the CSV clean without an extra index column
cohort.to_csv(os.path.join(csv_folder, 'cohort_labels.csv'), index=False)
print("\nCohort saved to cohort_labels.csv")
```

```python
# ============================================================
# STEP 2, BLOCK 11 (CORRECTED) — Pull NEUROBAT from the right visit
# ============================================================
# WHY THE FIX: RAVLT, Trails, and fluency are recorded at the
# 'bl' (baseline) visit. Logical Memory is recorded at 'sc'
# (screening). Taking the single earliest visit per patient
# grabbed the screening row and missed all the RAVLT data.
# The fix: for the battery tests, filter specifically to the
# baseline visit where that data actually lives.

neurobat = pd.read_csv(os.path.join(csv_folder, 'NEUROBAT.csv'),
                       low_memory=False)

# Filter to the baseline visit specifically, where RAVLT/Trails/
# fluency are recorded. We use VISCODE2 == 'bl'.
nb_bl = neurobat[neurobat['VISCODE2'] == 'bl'].copy()

# Safety: one row per patient at baseline
nb_bl = nb_bl.drop_duplicates(subset='RID', keep='first')
print(f"NEUROBAT baseline rows: {len(nb_bl):,}")

# --- Engineer the RAVLT total learning feature ---
ravlt_trials = ['AVTOT1', 'AVTOT2', 'AVTOT3', 'AVTOT4', 'AVTOT5']
nb_bl['RAVLT_learning_total'] = nb_bl[ravlt_trials].sum(
    axis=1, skipna=False
)

# --- Handle the Trail Making 999 ceiling (could-not-complete) ---
for col in ['TRAASCOR', 'TRABSCOR']:
    if col in nb_bl.columns:
        n_ceiling = (nb_bl[col] == 999).sum()
        nb_bl[col] = nb_bl[col].replace(999, np.nan)
        print(f"{col}: replaced {n_ceiling} ceiling (999) values with NaN")

# --- Select the battery features (recorded at baseline) ---
battery_features = [
    'RID', 'RAVLT_learning_total', 'AVDEL30MIN', 'AVDELTOT',
    'TRAASCOR', 'TRABSCOR', 'CATANIMSC', 'CATVEGESC',
]
nb_battery = nb_bl[[c for c in battery_features
                    if c in nb_bl.columns]].copy()

# --- Logical Memory lives at screening — grab it separately ---
# We take each patient's earliest non-missing Logical Memory
# value regardless of visit code, since it's a valid baseline-era
# measure recorded at screening.
lm = neurobat.dropna(subset=['LIMMTOTAL']).copy()
lm['VISDATE'] = pd.to_datetime(lm['VISDATE'], errors='coerce')
lm = lm.sort_values(['RID', 'VISDATE'])
lm = lm.drop_duplicates(subset='RID', keep='first')
lm_features = lm[['RID', 'LIMMTOTAL', 'LDELTOTAL']].copy()

# --- Merge both onto the master table ---
master = master.merge(nb_battery, on='RID', how='left')
master = master.merge(lm_features, on='RID', how='left')

print(f"\nMaster table now: {len(master):,} patients, "
      f"{len(master.columns)} columns")

# --- Check coverage ---
print(f"\nNEUROBAT feature coverage within your cohort:")
for col in ['RAVLT_learning_total', 'AVDEL30MIN', 'AVDELTOT',
            'TRAASCOR', 'TRABSCOR', 'CATANIMSC', 'CATVEGESC',
            'LIMMTOTAL', 'LDELTOTAL']:
    if col in master.columns:
        have = master[col].notna().sum()
        pct  = have / len(master) * 100
        print(f"  {col:22} {have:4} patients ({pct:.0f}%)  "
              f"— {100-pct:.0f}% missing")
```

```python
# ============================================================
# MASTER DATASET BUILD — clean, ordered, reproducible
# ============================================================
# This rebuilds the entire analysis dataset from scratch in one
# pass. Run after a kernel restart for a guaranteed-clean result.

import pandas as pd
import numpy as np
import os

csv_folder = r"C:\Users\Chima\Downloads\ADNIMERGE2\ADNIMERGE2\data_csv"

# ---- Reusable helper: earliest valid baseline row per patient ----
# Takes each patient's earliest record (optionally requiring a
# specific column to be non-missing, so we land on a real visit).
def earliest_per_patient(df, date_col, require_col=None):
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    df = df.dropna(subset=[date_col])
    if require_col:
        df = df.dropna(subset=[require_col])
    df = df.sort_values(['RID', date_col])
    return df.drop_duplicates(subset='RID', keep='first')

# ============================================================
# PART 1 — Load the labeled cohort (built earlier, saved to CSV)
# ============================================================
cohort = pd.read_csv(os.path.join(csv_folder, 'cohort_labels.csv'))
cohort = cohort[cohort['label'].notna()].copy()
cohort['baseline_date'] = pd.to_datetime(cohort['baseline_date'],
                                         errors='coerce')
master = cohort.copy()
print(f"Cohort: {len(master)} patients "
      f"({(master['label']==1).sum()} converters, "
      f"{(master['label']==0).sum()} stable)")

# ============================================================
# PART 2 — Cognitive features (earliest visit by date)
# ============================================================
adas = earliest_per_patient(
    pd.read_csv(os.path.join(csv_folder, 'ADAS.csv'), low_memory=False),
    'VISDATE')[['RID', 'TOTSCORE', 'TOTAL13']]

mmse = earliest_per_patient(
    pd.read_csv(os.path.join(csv_folder, 'MMSE.csv'), low_memory=False),
    'VISDATE')[['RID', 'MMSCORE']]

cdr = earliest_per_patient(
    pd.read_csv(os.path.join(csv_folder, 'CDR.csv'), low_memory=False),
    'VISDATE')[['RID', 'CDRSB', 'CDGLOBAL']]

faq = earliest_per_patient(
    pd.read_csv(os.path.join(csv_folder, 'FAQ.csv'), low_memory=False),
    'VISDATE')[['RID', 'FAQTOTAL']]

gds = earliest_per_patient(
    pd.read_csv(os.path.join(csv_folder, 'GDSCALE.csv'), low_memory=False),
    'VISDATE')[['RID', 'GDTOTAL']]

for df in [adas, mmse, cdr, faq, gds]:
    master = master.merge(df, on='RID', how='left')
print(f"After cognitive merge: {master.shape[1]} columns")

# ============================================================
# PART 3 — Neuropsychological battery (RAVLT/Trails at 'bl',
#          Logical Memory from earliest visit)
# ============================================================
neurobat = pd.read_csv(os.path.join(csv_folder, 'NEUROBAT.csv'),
                       low_memory=False)

# Battery tests live at baseline visit 'bl'
nb_bl = neurobat[neurobat['VISCODE2'] == 'bl'].drop_duplicates(
    subset='RID', keep='first').copy()
nb_bl['RAVLT_learning_total'] = nb_bl[
    ['AVTOT1', 'AVTOT2', 'AVTOT3', 'AVTOT4', 'AVTOT5']
].sum(axis=1, skipna=False)
for col in ['TRAASCOR', 'TRABSCOR']:
    nb_bl[col] = nb_bl[col].replace(999, np.nan)
nb_battery = nb_bl[['RID', 'RAVLT_learning_total', 'AVDEL30MIN',
                    'AVDELTOT', 'TRAASCOR', 'TRABSCOR', 'CATANIMSC']].copy()

# Logical Memory lives at screening — earliest non-missing
lm = earliest_per_patient(neurobat, 'VISDATE', require_col='LIMMTOTAL')
lm_feat = lm[['RID', 'LIMMTOTAL', 'LDELTOTAL']].copy()

master = master.merge(nb_battery, on='RID', how='left')
master = master.merge(lm_feat, on='RID', how='left')
print(f"After NEUROBAT merge: {master.shape[1]} columns")

# ============================================================
# PART 4 — Demographics + computed AGE + APOE4 count
# ============================================================
ptdemog = earliest_per_patient(
    pd.read_csv(os.path.join(csv_folder, 'PTDEMOG.csv'), low_memory=False),
    'VISDATE')[['RID', 'PTGENDER', 'PTDOBYY', 'PTEDUCAT',
                'PTRACCAT', 'PTETHCAT', 'PTMARRY']]
master = master.merge(ptdemog, on='RID', how='left')
master['AGE'] = master['baseline_date'].dt.year - master['PTDOBYY']

apoe = pd.read_csv(os.path.join(csv_folder, 'APOERES.csv'), low_memory=False)
apoe['APOE4_count'] = apoe['GENOTYPE'].astype(str).str.count('4')
apoe_clean = apoe[['RID', 'APOE4_count']].drop_duplicates(subset='RID')
master = master.merge(apoe_clean, on='RID', how='left')
print(f"After demographics + APOE merge: {master.shape[1]} columns")

# ============================================================
# PART 5 — MRI brain volumes (FSX7, earliest valid scan)
# ============================================================
fsx7 = pd.read_csv(os.path.join(csv_folder, 'UCSFFSX7.csv'),
                   low_memory=False)
fsx7_bl = earliest_per_patient(fsx7, 'EXAMDATE', require_col='ST29SV')
fsx7_bl['Hippocampus_total'] = fsx7_bl[['ST29SV','ST88SV']].sum(axis=1, skipna=False)
fsx7_bl['Entorhinal_total']  = fsx7_bl[['ST24CV','ST83CV']].sum(axis=1, skipna=False)
fsx7_bl['Ventricle_total']   = fsx7_bl[['ST37SV','ST96SV']].sum(axis=1, skipna=False)
fsx7_bl['Fusiform_total']    = fsx7_bl[['ST26CV','ST85CV']].sum(axis=1, skipna=False)
fsx7_bl['ICV']               = fsx7_bl['ST10CV']
mri = fsx7_bl[['RID', 'Hippocampus_total', 'Entorhinal_total',
               'Ventricle_total', 'Fusiform_total', 'ICV']].copy()
master = master.merge(mri, on='RID', how='left')
print(f"After MRI merge: {master.shape[1]} columns")

# ============================================================
# FINAL CHECK
# ============================================================
dupe_cols = master.columns[master.columns.duplicated()].tolist()
print(f"\nFinal master: {master.shape[0]} patients, {master.shape[1]} columns")
print(f"Duplicated columns: {dupe_cols if dupe_cols else 'None — clean'}")
```

```python
# ============================================================
# STEP 5, BLOCK 19 — Merge CSF biomarkers (final feature merge)
# ============================================================
# WHY: CSF amyloid and tau are powerful biological markers but
# only collected via lumbar puncture in earlier ADNI phases, so
# coverage will be partial. We use the harmonized (non-RAW)
# values, which are adjusted to be comparable across the
# different lab batches and kits used over the years.

upenn = pd.read_csv(os.path.join(csv_folder, 'UPENNBIOMK_MASTER.csv'),
                    low_memory=False)

# This file uses DRAWDTE (the date CSF was drawn) as its date.
# Take each patient's EARLIEST valid measurement (one that has
# an actual ABETA value, so we land on a real assay record).
upenn['DRAWDTE'] = pd.to_datetime(upenn['DRAWDTE'], errors='coerce')
upenn_valid = upenn.dropna(subset=['ABETA']).copy()
upenn_valid = upenn_valid.sort_values(['RID', 'DRAWDTE'])
csf = upenn_valid.drop_duplicates(subset='RID', keep='first')

# Keep the three harmonized biomarkers (not the _RAW versions)
csf_features = csf[['RID', 'ABETA', 'TAU', 'PTAU']].copy()
print(f"Patients with CSF data: {len(csf_features):,}")

# Merge onto master
master = master.merge(csf_features, on='RID', how='left')

print(f"\nMaster table now: {len(master):,} patients, "
      f"{len(master.columns)} columns")

# --- Coverage check ---
print(f"\nCSF biomarker coverage within cohort:")
for col in ['ABETA', 'TAU', 'PTAU']:
    have = master[col].notna().sum()
    pct = have / len(master) * 100
    print(f"  {col:8} {have:4} patients ({pct:.0f}%)  "
          f"— {100-pct:.0f}% missing")
```

```python
# ============================================================
# STEP 6, BLOCK 20 — Organize columns & define feature sets
# ============================================================
# WHY: Before modeling we must clearly separate (a) the label
# we predict, (b) metadata that must NEVER enter the model, and
# (c) the actual predictor features. Mixing these up causes
# data leakage, so we are explicit about each group.

# --- The outcome we predict ---
LABEL = 'label'

# --- Metadata / identifier columns — NOT features ---
# These either identify patients or directly encode the outcome.
# conversion_month and max_followup_months would leak the answer,
# so they are strictly excluded from the predictor set.
NON_FEATURES = ['RID', 'PTID', 'baseline_date', 'label',
                'conversion_month', 'max_followup_months',
                'PTDOBYY']  # birth year — we already derived AGE from it

# --- The CSF biomarker columns (define the full vs minimal split) ---
CSF_FEATURES = ['ABETA', 'TAU', 'PTAU']

# --- Build the MINIMAL-resource feature list ---
# Everything that is NOT metadata and NOT CSF
minimal_features = [c for c in master.columns
                    if c not in NON_FEATURES and c not in CSF_FEATURES]

# --- Build the FULL feature list ---
# The minimal features PLUS the CSF biomarkers
full_features = minimal_features + CSF_FEATURES

print("MINIMAL-resource feature set ({} features):".format(len(minimal_features)))
print(minimal_features)
print("\nFULL feature set ({} features):".format(len(full_features)))
print(full_features)
```

```python
# ============================================================
# STEP 6, BLOCK 22 — Inspect categorical variables before encoding
# ============================================================
# WHY: Race, ethnicity, sex, and marital status are stored as
# numeric codes that represent CATEGORIES, not quantities. Before
# encoding we need to see how many categories each has and how
# patients are distributed, which determines the encoding approach.

for col in ['PTGENDER', 'PTRACCAT', 'PTETHCAT', 'PTMARRY']:
    print(f"\n{col} value distribution:")
    print(master[col].value_counts(dropna=False).sort_index())
```

```python
# ============================================================
# STEP 6, BLOCK 23 — Encode categorical variables
# ============================================================
# WHY: Models need numbers, and categories with tiny counts
# create useless columns. Given ADNI's homogeneity (93% White),
# we collapse categories into meaningful, well-populated groups
# rather than one-hot encoding tiny categories. Each encoding
# choice is documented because it has analytic consequences.

# --- Sex: simple binary (Female=1, Male=0) ---
master['sex_female'] = (master['PTGENDER'] == 'Female').astype(int)

# --- Race: collapse to White vs Non-White ---
# WHY: minority categories have 1-29 patients each, too few for
# separate columns. White vs Non-White is the only split with
# enough data to be meaningful. Finer analysis deferred to NACC.
master['race_nonwhite'] = (master['PTRACCAT'] != 'White').astype(int)

# --- Ethnicity: Hispanic vs Not ---
master['ethnicity_hispanic'] = (
    master['PTETHCAT'] == 'Hispanic or Latino').astype(int)

# --- Marital status: Married vs Not-Married ---
master['married'] = (master['PTMARRY'] == 'Married').astype(int)

# --- Confirm the new encoded columns ---
print("Encoded categorical features created:")
for col in ['sex_female', 'race_nonwhite',
            'ethnicity_hispanic', 'married']:
    print(f"\n{col}:")
    print(master[col].value_counts().sort_index())
```

```python
# ============================================================
# STEP 6, BLOCK 24 — Finalize feature lists with encoded columns
# ============================================================
# WHY: We replace the original TEXT categorical columns
# (PTGENDER, PTRACCAT, PTETHCAT, PTMARRY) with the new NUMERIC
# encoded versions (sex_female, race_nonwhite, etc.), since
# models need numbers. We rebuild the feature lists cleanly.

# Columns that are NOT features (metadata, identifiers, label,
# AND the original text categoricals now replaced by encodings)
NON_FEATURES = ['RID', 'PTID', 'baseline_date', 'label',
                'conversion_month', 'max_followup_months',
                'PTDOBYY',
                # original text categoricals — replaced by encoded versions
                'PTGENDER', 'PTRACCAT', 'PTETHCAT', 'PTMARRY']

CSF_FEATURES = ['ABETA', 'TAU', 'PTAU']

# Rebuild the minimal feature list: everything that is not
# metadata and not CSF (this now includes the encoded columns)
minimal_features = [c for c in master.columns
                    if c not in NON_FEATURES and c not in CSF_FEATURES]

full_features = minimal_features + CSF_FEATURES

print(f"MINIMAL feature set ({len(minimal_features)} features):")
for f in minimal_features:
    print(f"  - {f}")
print(f"\nFULL feature set: {len(full_features)} features "
      f"(minimal + ABETA, TAU, PTAU)")
```

```python
# ============================================================
# STEP 6, BLOCK 25 — Check if CDGLOBAL is informative
# ============================================================
# WHY: MCI is partly DEFINED by a global CDR of 0.5. If nearly
# all your patients share the same CDGLOBAL value, the feature
# carries almost no information to distinguish converters from
# non-converters, and may even subtly encode the diagnosis
# criterion itself. We check its distribution before deciding.

print("CDGLOBAL distribution across the cohort:")
print(master['CDGLOBAL'].value_counts(dropna=False).sort_index())

# Also see how it differs between converters and non-converters
print("\nCDGLOBAL by outcome (mean):")
print(master.groupby('label')['CDGLOBAL'].mean())
```

```python
# ============================================================
# STEP 6, BLOCK 26 — Drop CDGLOBAL, finalize feature lists
# ============================================================
# WHY: CDGLOBAL is near-constant (98.8% = 0.5) and nearly
# identical between converters and non-converters, so it carries
# no predictive signal. It also partly restates the MCI inclusion
# criterion, risking circularity. We exclude it. CDRSB (the Sum
# of Boxes) is retained because it has genuine spread and signal.

# Add CDGLOBAL to the non-features list
NON_FEATURES = ['RID', 'PTID', 'baseline_date', 'label',
                'conversion_month', 'max_followup_months',
                'PTDOBYY',
                'PTGENDER', 'PTRACCAT', 'PTETHCAT', 'PTMARRY',
                'CDGLOBAL']  # near-constant, dropped

CSF_FEATURES = ['ABETA', 'TAU', 'PTAU']

# Rebuild the feature lists
minimal_features = [c for c in master.columns
                    if c not in NON_FEATURES and c not in CSF_FEATURES]
full_features = minimal_features + CSF_FEATURES

print(f"FINAL minimal feature set ({len(minimal_features)} features):")
for f in minimal_features:
    print(f"  - {f}")
print(f"\nFINAL full feature set: {len(full_features)} features")
```

```python
# ============================================================
# STEP 7, BLOCK 27 — Train/test split (both feature sets)
# ============================================================
# WHY: We hold back 20% of patients the model never sees during
# training, to honestly estimate real-world performance. We
# stratify by the label so the 71/29 class balance is preserved
# in both sets, and we fix a random seed for reproducibility.

from sklearn.model_selection import train_test_split

# Our outcome variable
y = master['label'].astype(int)

# --- MINIMAL feature set: all 816 patients ---
X_min = master[minimal_features].copy()

# --- FULL feature set: only patients WITH CSF data ---
# The full model can only use patients who have biomarkers, so
# we restrict to rows where ABETA is present. This is the ~540
# patient subset. The minimal model keeps all 816.
has_csf = master['ABETA'].notna()
X_full = master.loc[has_csf, full_features].copy()
y_full = master.loc[has_csf, 'label'].astype(int)

# --- Split the MINIMAL set (all 816) ---
# test_size=0.2 -> 20% held out; stratify=y keeps class balance;
# random_state=42 makes the split reproducible
X_min_train, X_min_test, y_min_train, y_min_test = train_test_split(
    X_min, y, test_size=0.2, stratify=y, random_state=42
)

# --- Split the FULL set (CSF subset) ---
X_full_train, X_full_test, y_full_train, y_full_test = train_test_split(
    X_full, y_full, test_size=0.2, stratify=y_full, random_state=42
)

# --- Report the splits ---
print("MINIMAL model data (all patients):")
print(f"  Training: {len(X_min_train)} patients "
      f"({y_min_train.sum()} converters, {(y_min_train==0).sum()} stable)")
print(f"  Testing:  {len(X_min_test)} patients "
      f"({y_min_test.sum()} converters, {(y_min_test==0).sum()} stable)")

print("\nFULL model data (CSF subset):")
print(f"  Training: {len(X_full_train)} patients "
      f"({y_full_train.sum()} converters, {(y_full_train==0).sum()} stable)")
print(f"  Testing:  {len(X_full_test)} patients "
      f"({y_full_test.sum()} converters, {(y_full_test==0).sum()} stable)")
```

```python
# ============================================================
# STEP 8, BLOCK 28 — First model: logistic regression (minimal)
# ============================================================
# WHY: Logistic regression is our interpretable baseline. We
# wrap imputation + scaling + the model in a Pipeline so the
# imputation medians and scaling are learned ONLY from training
# data, then applied to test data — preventing leakage by design.

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

# Build the pipeline: 3 steps run in order
pipe_lr = Pipeline([
    # Step 1: fill missing values with the median (learned from
    # training data only when we call .fit)
    ('impute', SimpleImputer(strategy='median')),

    # Step 2: standardize features to comparable scales so no
    # feature dominates just because its numbers are larger
    # (e.g. ICV in the thousands vs MMSE 0-30)
    ('scale', StandardScaler()),

    # Step 3: the model. class_weight='balanced' tells it to pay
    # proportionally more attention to the minority class
    # (converters), addressing our 71/29 imbalance.
    # max_iter raised so it has enough iterations to converge.
    ('model', LogisticRegression(class_weight='balanced',
                                 max_iter=1000, random_state=42))
])

# Train the pipeline on the MINIMAL training data
# .fit learns imputation medians, scaling, AND model weights —
# all from training data only
pipe_lr.fit(X_min_train, y_min_train)

# --- Evaluate on the held-out test set ---
# predict_proba gives the probability of conversion (class 1)
# [:, 1] selects the probability of the positive class
y_pred_proba = pipe_lr.predict_proba(X_min_test)[:, 1]
y_pred       = pipe_lr.predict(X_min_test)

# AUC-ROC: our primary metric. Ranges 0.5 (random) to 1.0
# (perfect). Measures how well the model ranks converters above
# non-converters. Robust to class imbalance.
auc = roc_auc_score(y_min_test, y_pred_proba)

print(f"{'='*50}")
print(f"LOGISTIC REGRESSION — Minimal feature set")
print(f"{'='*50}")
print(f"\nAUC-ROC on held-out test set: {auc:.3f}")
print(f"\nClassification report:")
print(classification_report(y_min_test, y_pred,
                            target_names=['Stable', 'Converter']))
print(f"Confusion matrix:")
print(f"(rows = actual, columns = predicted)")
print(confusion_matrix(y_min_test, y_pred))
```

```python
# ============================================================
# STEP 8, BLOCK 29 — Random forest & XGBoost (minimal set)
# ============================================================
# WHY: Tree-based models can capture feature interactions and
# nonlinear patterns that logistic regression misses. We use the
# same Pipeline pattern (impute -> scale -> model) so evaluation
# stays leakage-free and directly comparable to our LR baseline.

from sklearn.ensemble import RandomForestClassifier

# Install xgboost if needed (run once; safe to leave in)
try:
    from xgboost import XGBClassifier
except ImportError:
    import subprocess, sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "xgboost"])
    from xgboost import XGBClassifier

# --- RANDOM FOREST pipeline ---
# n_estimators=300: number of trees. class_weight='balanced'
# handles the 71/29 imbalance. Note: tree models don't strictly
# need scaling, but keeping it in the pipeline is harmless and
# keeps all three models consistent.
pipe_rf = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', RandomForestClassifier(
        n_estimators=300, class_weight='balanced',
        random_state=42, n_jobs=-1))
])

# --- XGBOOST pipeline ---
# scale_pos_weight handles imbalance: ratio of negatives to
# positives in training (~461/191 ≈ 2.4), telling XGBoost to
# weight the minority converter class more heavily.
neg = (y_min_train == 0).sum()
pos = (y_min_train == 1).sum()
scale_pos = neg / pos

pipe_xgb = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', XGBClassifier(
        n_estimators=300, max_depth=4, learning_rate=0.05,
        scale_pos_weight=scale_pos, eval_metric='logloss',
        random_state=42))
])

# --- Train and evaluate both, reusing our LR result ---
results = {'Logistic Regression': 0.888}  # from Block 28

for name, pipe in [('Random Forest', pipe_rf), ('XGBoost', pipe_xgb)]:
    pipe.fit(X_min_train, y_min_train)
    proba = pipe.predict_proba(X_min_test)[:, 1]
    auc = roc_auc_score(y_min_test, proba)
    results[name] = auc
    print(f"\n{'='*50}")
    print(f"{name} — Minimal feature set")
    print(f"{'='*50}")
    print(f"AUC-ROC: {auc:.3f}")
    pred = pipe.predict(X_min_test)
    print(classification_report(y_min_test, pred,
                                target_names=['Stable', 'Converter']))

# --- Three-way comparison ---
print(f"\n{'='*50}")
print(f"MODEL COMPARISON (Minimal feature set, AUC-ROC)")
print(f"{'='*50}")
for name, auc in sorted(results.items(), key=lambda x: -x[1]):
    print(f"  {name:22} {auc:.3f}")
```

```python
# ============================================================
# STEP 8, BLOCK 30 — All three models on the FULL feature set
# ============================================================
# WHY: Tests the core tradeoff — does adding CSF biomarkers
# (richer features) outweigh the loss of patients (540 vs 816)?
# We run the same three pipelines on the full-feature CSF subset
# and compare against the minimal benchmark of 0.888.

# Recompute scale_pos_weight for the full training set
neg_f = (y_full_train == 0).sum()
pos_f = (y_full_train == 1).sum()
scale_pos_f = neg_f / pos_f

full_models = {
    'Logistic Regression': Pipeline([
        ('impute', SimpleImputer(strategy='median')),
        ('scale', StandardScaler()),
        ('model', LogisticRegression(class_weight='balanced',
                                     max_iter=1000, random_state=42))]),
    'Random Forest': Pipeline([
        ('impute', SimpleImputer(strategy='median')),
        ('scale', StandardScaler()),
        ('model', RandomForestClassifier(n_estimators=300,
                  class_weight='balanced', random_state=42, n_jobs=-1))]),
    'XGBoost': Pipeline([
        ('impute', SimpleImputer(strategy='median')),
        ('scale', StandardScaler()),
        ('model', XGBClassifier(n_estimators=300, max_depth=4,
                  learning_rate=0.05, scale_pos_weight=scale_pos_f,
                  eval_metric='logloss', random_state=42))])
}

full_results = {}
for name, pipe in full_models.items():
    pipe.fit(X_full_train, y_full_train)
    proba = pipe.predict_proba(X_full_test)[:, 1]
    auc = roc_auc_score(y_full_test, proba)
    full_results[name] = auc
    pred = pipe.predict(X_full_test)
    print(f"\n{'='*50}")
    print(f"{name} — FULL feature set (with CSF, n={len(X_full_train)} train)")
    print(f"{'='*50}")
    print(f"AUC-ROC: {auc:.3f}")
    print(classification_report(y_full_test, pred,
                                target_names=['Stable', 'Converter']))

# --- The key comparison ---
print(f"\n{'='*50}")
print(f"MINIMAL vs FULL — does CSF help?")
print(f"{'='*50}")
print(f"  Minimal (no CSF, 816 patients):  LR = 0.888")
print(f"  Full (with CSF, 540 patients):")
for name, auc in sorted(full_results.items(), key=lambda x: -x[1]):
    print(f"      {name:20} {auc:.3f}")
```

```python
# ============================================================
# STEP 8, BLOCK 31 — Apples-to-apples: same patients, both models
# ============================================================
# WHY: The minimal vs full comparison used different test sets
# (164 vs 109 patients). To isolate the effect of CSF features
# alone, we evaluate BOTH models on the SAME patients — the
# full model's test set (the CSF subset). Any performance
# difference is then due purely to the features, not the people.

# The full model's test patients (the fair common ground)
common_test_idx = X_full_test.index

# Their MINIMAL features (same patients, without CSF)
X_common_min = master.loc[common_test_idx, minimal_features]
# Their FULL features (same patients, with CSF)
X_common_full = master.loc[common_test_idx, full_features]
# Their true labels
y_common = master.loc[common_test_idx, 'label'].astype(int)

# Retrain the minimal champion (logistic regression) on the
# MINIMAL training data, then test on these common patients
pipe_lr_min = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', LogisticRegression(class_weight='balanced',
                                 max_iter=1000, random_state=42))])
pipe_lr_min.fit(X_min_train, y_min_train)

# Retrain the full champion (random forest) on FULL training data
pipe_rf_full = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', RandomForestClassifier(n_estimators=300,
              class_weight='balanced', random_state=42, n_jobs=-1))])
pipe_rf_full.fit(X_full_train, y_full_train)

# Evaluate BOTH on the SAME common test patients
auc_min_common = roc_auc_score(
    y_common, pipe_lr_min.predict_proba(X_common_min)[:, 1])
auc_full_common = roc_auc_score(
    y_common, pipe_rf_full.predict_proba(X_common_full)[:, 1])

print(f"{'='*55}")
print(f"APPLES-TO-APPLES — same {len(y_common)} patients, both models")
print(f"{'='*55}")
print(f"  Minimal model (LR, no CSF):      AUC = {auc_min_common:.3f}")
print(f"  Full model (RF, with CSF):       AUC = {auc_full_common:.3f}")
print(f"  Difference attributable to CSF:  {auc_full_common - auc_min_common:+.3f}")
```

```python
# ============================================================
# STEP 8, BLOCK 32 — Isolate CSF effect: same algorithm, same patients
# ============================================================
# WHY: The previous comparison changed BOTH the features AND the
# algorithm (LR vs RF). To isolate the pure effect of CSF, we use
# the SAME algorithm (logistic regression) with and without CSF,
# on the SAME 109 common patients.

# Logistic regression WITHOUT CSF (minimal), already trained:
auc_lr_nocsf = roc_auc_score(
    y_common, pipe_lr_min.predict_proba(X_common_min)[:, 1])

# Logistic regression WITH CSF (full), trained on full training data
pipe_lr_full = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', LogisticRegression(class_weight='balanced',
                                 max_iter=1000, random_state=42))])
pipe_lr_full.fit(X_full_train, y_full_train)
auc_lr_csf = roc_auc_score(
    y_common, pipe_lr_full.predict_proba(X_common_full)[:, 1])

print(f"{'='*55}")
print(f"PURE CSF EFFECT — same algorithm (LR), same {len(y_common)} patients")
print(f"{'='*55}")
print(f"  LR without CSF:  AUC = {auc_lr_nocsf:.3f}")
print(f"  LR with CSF:     AUC = {auc_lr_csf:.3f}")
print(f"  Pure CSF effect: {auc_lr_csf - auc_lr_nocsf:+.3f}")
```

```python
# ============================================================
# STEP 9, BLOCK 33 — Cross-validation for stable AUC estimates
# ============================================================
# WHY: A single train/test split can be lucky or unlucky. 5-fold
# cross-validation trains and tests on 5 different rotations of
# the data, giving an averaged AUC plus its spread (standard
# deviation). This tells us how STABLE our performance really is.
# StratifiedKFold preserves the 71/29 class balance in each fold.

from sklearn.model_selection import cross_val_score, StratifiedKFold

# 5 folds, stratified to preserve class balance, shuffled with seed
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# --- Cross-validate the MINIMAL champion (logistic regression) ---
# We use the FULL minimal dataset (all 816 patients) here, since
# CV handles the splitting internally.
pipe_lr_cv = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', LogisticRegression(class_weight='balanced',
                                 max_iter=1000, random_state=42))])

# scoring='roc_auc' computes AUC on each fold
min_scores = cross_val_score(pipe_lr_cv, X_min, y,
                             cv=cv, scoring='roc_auc')

print("MINIMAL model (Logistic Regression) — 5-fold CV:")
print(f"  AUC per fold: {[f'{s:.3f}' for s in min_scores]}")
print(f"  Mean AUC: {min_scores.mean():.3f} "
      f"(+/- {min_scores.std():.3f})")

# --- Cross-validate the FULL champion (random forest) ---
pipe_rf_cv = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler()),
    ('model', RandomForestClassifier(n_estimators=300,
              class_weight='balanced', random_state=42, n_jobs=-1))])

full_scores = cross_val_score(pipe_rf_cv, X_full, y_full,
                              cv=cv, scoring='roc_auc')

print("\nFULL model (Random Forest) — 5-fold CV:")
print(f"  AUC per fold: {[f'{s:.3f}' for s in full_scores]}")
print(f"  Mean AUC: {full_scores.mean():.3f} "
      f"(+/- {full_scores.std():.3f})")
```

```python
# ============================================================
# Bootstrap 95% confidence intervals for AUC
# ============================================================
# WHY: A single AUC is a point estimate. Bootstrapping resamples
# the test set many times to estimate the range the true AUC
# plausibly falls in. We compute it for both champion models on
# their held-out test sets.

from sklearn.metrics import roc_auc_score
import numpy as np

def bootstrap_auc_ci(model, X_test, y_test, n_boot=2000, seed=42):
    """Resample the test set n_boot times, computing AUC each time,
    then return the median and 95% confidence interval."""
    rng = np.random.RandomState(seed)
    # The model's predicted probabilities on the test set (fixed)
    proba = model.predict_proba(X_test)[:, 1]
    y_arr = np.array(y_test)
    n = len(y_arr)

    aucs = []
    for _ in range(n_boot):
        # Resample patient indices WITH replacement
        idx = rng.randint(0, n, n)
        # Skip resamples that accidentally contain only one class
        if len(np.unique(y_arr[idx])) < 2:
            continue
        aucs.append(roc_auc_score(y_arr[idx], proba[idx]))

    aucs = np.array(aucs)
    lower = np.percentile(aucs, 2.5)
    upper = np.percentile(aucs, 97.5)
    return np.median(aucs), lower, upper

# --- Minimal model (logistic regression) on its test set ---
# Refit on the minimal training split for an honest test evaluation
pipe_lr_min.fit(X_min_train, y_min_train)
med_min, lo_min, hi_min = bootstrap_auc_ci(
    pipe_lr_min, X_min_test, y_min_test)

# --- Full model (random forest) on its test set ---
pipe_rf_full.fit(X_full_train, y_full_train)
med_full, lo_full, hi_full = bootstrap_auc_ci(
    pipe_rf_full, X_full_test, y_full_test)

print("="*55)
print("BOOTSTRAP 95% CONFIDENCE INTERVALS (2000 resamples)")
print("="*55)
print(f"\nMINIMAL model (LR, n_test={len(y_min_test)}):")
print(f"  AUC median: {med_min:.3f}")
print(f"  95% CI: [{lo_min:.3f}, {hi_min:.3f}]")
print(f"  CI width: {hi_min - lo_min:.3f}")

print(f"\nFULL model (RF, n_test={len(y_full_test)}):")
print(f"  AUC median: {med_full:.3f}")
print(f"  95% CI: [{lo_full:.3f}, {hi_full:.3f}]")
print(f"  CI width: {hi_full - lo_full:.3f}")
```
